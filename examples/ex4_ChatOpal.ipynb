````python
import os
os.environ["OPENAI_API_KEY"] = "sk-xxxxxxxxxxxxxxx"
os.environ["OPENLINK_API_KEY"] = "xxxxxxxxxxxxxxxx"


from langchain_opal import ChatOpal

# print list of Models
lst = ChatOpal.get_models_list()
for x in lst:
    print(f"id= {x['id']:40}  \t name= {x['name']}")
````

````python
# print list of FineTunes
lst = ChatOpal.get_finetunes_list()
for x in lst:
    print(f"id= {x['chat_id']:40} \t title= {x['title']}")
````

````python
# print list of Functions
lst = ChatOpal.get_functions_list()
for x in lst:
    print(f"function= {x['function']:30} \t title= {x['title']}")
````

````python
llm = ChatOpal()
messages = [
        ("system", "You are a helpful translator. Translate the user sentence to French."),
        ("human", "I love programming."),
    ]
resp = llm.invoke(messages)
print(resp)
print("----------------------------------")
print(resp.content)
````

````python
llm = ChatOpal(
    model_name="gpt-4o",
    temperature=0.2,
    top_p=0.5,
    api_base="https://linkeddata.uriburner.com",
    finetune="system-data-twingler-config",
    funcs_list=["UB.DBA.sparqlQuery", "DB.DBA.vos_howto_search", "Demo.demo.execute_sql_query", "DB.DBA.graphqlQuery"],
)
messages = [
        ("system", "You are a helpful assistant."),
        ("human", "Give me list of Movies by Spike."),
    ]
resp = llm.invoke(messages)
print(resp)
print("----------------------------------")
print(resp.content)
````
